# Creating a Sample Dataset

To demonstrate different types of visualizations, a simple student performance dataset is created. This dataset includes students' marks, study hours, attendance, and participation in extracurricular activities.

```python
import pandas as pd

data = {
    "Student": ["A", "B", "C", "D", "E", "F"],
    "Marks": [85, 92, 78, 88, 95, 80],
    "Study_Hours": [2, 4, 3, 5, 6, 3],
    "Attendance": [90, 95, 85, 92, 98, 88],
    "Activity": ["Yes", "Yes", "No", "Yes", "No", "Yes"]
}

df = pd.DataFrame(data)

df
```

# 5. Matplotlib Graphs

## 5.1 Line Plot

### Description

A line plot displays data points connected by straight lines. It is commonly used to show changes or trends over a period or across ordered categories.

### Use Case

Line plots are useful for tracking values such as monthly sales, stock prices, temperatures, or student performance over time.

```python
plt.figure(figsize=(6,4))
plt.plot(df["Student"], df["Marks"],marker="o",color="blue")
plt.title("Student Marks")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

### Observation

The line plot shows the marks scored by each student. Student E achieved the highest marks, while Student C scored the lowest. The graph makes it easy to compare the performance of all students.

```python
plt.figure(figsize=(6,4))

plt.plot(df["Student"], df["Marks"], marker="o")

plt.title("Student Marks")

plt.xlabel("Students")

plt.ylabel("Marks")

plt.grid(True)

plt.savefig("line_plot.png")

plt.show()
```

## 5.2 Scatter Plot

### Description

A scatter plot displays individual data points on a graph to show the relationship between two numerical variables. It helps identify patterns, trends, clusters, or outliers within the data.

### Use Case

Scatter plots are commonly used to analyze relationships between variables, such as study hours and marks, height and weight, or advertising expenditure and sales.

```python
plt.figure(figsize=(6,4))

plt.scatter(df["Study_Hours"], df["Marks"],
            color="green",
            s=100)

plt.title("Study Hours vs Marks")

plt.xlabel("Study_Hours")

plt.ylabel("Marks")

plt.grid(True)

plt.show()
```

### Observation

The scatter plot shows the relationship between study hours and marks. Students who studied for more hours generally achieved higher marks. Although the dataset is small, the graph indicates a positive relationship between study time and academic performance.

## 5.3 Bar Chart

### Description

A bar chart represents data using rectangular bars, where the height of each bar corresponds to its value. It is one of the most effective ways to compare values across different categories.

### Use Case

Bar charts are commonly used to compare sales of products, students' marks, monthly expenses, or any data that belongs to different categories.

```python
plt.figure(figsize=(6,4))

plt.bar(df["Student"], df["Marks"],
        color="orange")

plt.title("Student Marks")

plt.xlabel("Student")

plt.ylabel("Marks")

plt.show()
```

### Observation

The bar chart compares the marks of all students. It clearly shows that Student E scored the highest marks, while Student C scored the lowest. The differences between students' performances are easy to identify through the height of the bars.

## 5.4 Histogram

### Description

A histogram is used to display the distribution of numerical data by grouping values into intervals, called bins. It helps understand how the data is spread and whether there are any patterns or concentrations.

### Use Case

Histograms are commonly used to analyze the distribution of marks, ages, salaries, heights, or any continuous numerical data.

```python
plt.figure(figsize=(6,4))

plt.hist(df["Marks"],
         bins=5,
         color="orange",
         edgecolor="black")

plt.title("Distribution of Student Marks")

plt.xlabel("Marks")

plt.ylabel("Frequency")

plt.show()
```

### Observation

The histogram shows how students' marks are distributed across different ranges. Most students scored between **80 and 95 marks**, indicating that the overall performance of the class is good.

## 5.5 Pie Chart

### Description

A pie chart represents data as slices of a circle, where each slice shows the proportion of a category relative to the whole. It is useful for displaying percentages or shares.

### Use Case

Pie charts are commonly used to show budget allocation, market share, survey results, or the percentage contribution of different categories.

```python
plt.figure(figsize=(6,6))

plt.pie(df["Marks"],
        labels=df["Student"],
        autopct="%1.1f%%",
        startangle=90)

plt.title("Percentage of Marks by Student")
```

### Observation

The pie chart shows each student's contribution to the total marks. Student E occupies the largest portion of the chart because they scored the highest marks, while Student C has the smallest share.

## 5.6 Box Plot

### Description

A box plot is used to summarize the distribution of numerical data. It displays the minimum value, first quartile (Q1), median, third quartile (Q3), and maximum value. It is also useful for identifying outliers.

### Use Case

Box plots are commonly used to compare the spread of data, detect outliers, and understand the distribution of marks, salaries, or other numerical values.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.boxplot(df["Marks"], patch_artist=True)

plt.title("Box Plot of Student Marks")
plt.ylabel("Marks")

plt.show()
```

#Observation

The box plot displays the spread of students' marks. The line inside the box represents the median, while the box shows the middle 50% of the data. Since there are no points outside the whiskers, there are no significant outliers in this dataset.

#5.7 Area Chart
### Description

An area chart is similar to a line chart, but the area below the line is filled with color. It helps highlight trends and the magnitude of values over a period.


### Use Case

Area charts are commonly used to visualize changes over time, such as sales, website traffic, rainfall, or student performance.

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

plt.fill_between(
    df["Student"],
    df["Marks"],
    color="skyblue",
    alpha=0.5
)

plt.plot(
    df["Student"],
    df["Marks"],
    marker="o",
    color="blue"
)

plt.title("Area Chart of Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.show()
```

### Observation

The area chart shows the marks scored by each student while highlighting the area beneath the line. It makes it easy to observe the overall trend, with Student E achieving the highest marks.

#6.6 Heatmap
### Description

A heatmap uses different colors to represent values in a matrix. It is commonly used to visualize correlations between numerical variables.
### Use Case

Heatmaps help identify positive or negative relationships between different variables in a dataset.

```python
df.head()
```

```python
plt.figure(figsize=(5,4))

correlation = df[["Marks", "Study_Hours", "Attendance"]].corr()

sns.heatmap(
    correlation,
    annot=True,
    cmap="Blues"
)

plt.title("Correlation Heatmap")

plt.show()
```

### Observation

The heatmap shows that study hours and attendance are positively related to students' marks.



# 7. Comparison Between Matplotlib and Seaborn


### Matplotlib

**Advantages:**

* Highly customizable
* Supports many graph types
* Works well with NumPy and Pandas

### Seaborn

Seaborn is built on top of Matplotlib and makes it easier to create attractive statistical graphs. It offers better default styles and requires less code.

**Advantages:**

* Easy to use
* Attractive default themes
* Excellent for statistical plots

### Overall Comparison

Matplotlib is best for creating customized visualizations, while Seaborn is better for quickly creating clean and informative statistical graphs. Both libraries are widely used together in data science projects.
